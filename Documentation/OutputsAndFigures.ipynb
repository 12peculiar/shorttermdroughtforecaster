# Output Methods

## Future projections

#### 900 daily future projections are run through the ecohydrology model. This data is synthesized by:

- First, future monthly variable means are calculated for each simulation.
- From these monthly means, the median, and the 10th and 90th quantile are calculated across all simulations.
- Historical data affiliation:
 - The affiliation of the historical climatology varies whether we are looking at forecast for the current year (1981 - 2010) or for upcoming year (1982 - 2011).
 - In order to calculate the difference between future monthly predictions and their relative historical climatologies, we calculate historical monthly values based on the correct reference climatologies.
   - First we find yearly monthly means
   - Then we calculate the median of these monthly means.
   - This data is not displayed.
- The difference between monthly future and historical medians, as well as the future 10th and 90th and the historical median are taken.
- Future monthly values are set the the midpoint (the 15th) of each month and interpolated for each day.

### Historical Climatology (black and gray)

The historical climatology varies whether we are loooking at prediction for the rest of the current year (i.e. 1981 - 2010) or
for the next year (i.e. 1982 - 2011). For the time period preceeding today (6 months), the entire historical climatological record is shown (1981 - 2011).

#### Top Panel 

The black line and gray shaded area are the historical climatology for the site. The black line is the median and the gray shaded area is the 10th - 90th quantile.

The historical climatology varies whether we are loooking at prediction for the rest of the current year (i.e. 1981 - 2010) or
for the next year (i.e. 1982 - 2011). For the time period preceeding today, all of the historical climatology is shown (1981 - 2011).

#### Bottom Panel

In the bottom panel the historical climatological median is set to 0, and the 10th to 90th quantiles are the differences between these values and 0 .

### Observed Data  (yellow)

Observed data is shown for the previous 6 months.

#### Top Panel

For the time period preceeding today, we see a 30-day rolling mean of observed dailys as a thick yellow line. Actual values are a thin yellow line. 

#### Bottom panel

To calculate the deltas in the bottom panel the difference between the thick yellow line, and the historical climatology (1981 - 2011) was taken.

If  soil moisture was less than the historical climatology it is shown in brown. If soil moisture was more than the historical climatology it is shown in green.


### Future Forecast Data (teal)

Future forecast data is shown for the upcoming 12 months.

#### Top Panel

The median of monthly means, as well as the 10th and 90th quantiles are shown in teal.

These values, were calculated as monthly values and interpolated to dailys. A full description is above.

#### Bottom Panel

To calculate the deltas in the bottom panel the difference between the thick and thin teal line, and the relative historical climatology (1981 - 2010 for the current year, 1982 - 2011 for next year) median was taken. 


If  soil moisture is predicted to be  less than the historical climatology it is shown in brown. If soil moisture is predicted to be more than the historical climatology it is shown in green.
